```python
# Count the number of missing values in each column
print(df2.isna().sum())
```

```python
# Count the number of missing values in each column
print(df2.isna().sum())

# Find the five percent threshold
threshold = len(df2) * 0.05

# Create a filter
cols_to_drop = df2.columns[df2.isna().sum() <= threshold]

# Drop missing values for columns below the threshold
df2.dropna(subset=cols_to_drop, inplace=True)

print(df2.isna().sum())
```

```python
# Check the values of the Additional_Info column
print(df2["Additional_Info"].value_counts())
```

```python
# Create a box plot of Price by Airline
sns.boxplot(data=df2, x='Price', y='Airline')

plt.show()
```

```python
# Calculate median plane ticket prices by Airline
airline_prices = df2.groupby("Airline")["Price"].median()

print(airline_prices)
```

```python
# Convert to a dictionary
prices_dict = airline_prices.to_dict()

# Map the dictionary to the missing values
df2["Price"] = df2["Price"].fillna(df2["Airline"].map(prices_dict))

# Check for missing values
print(df2.isna().sum())
```

```python
# Create a list of categories
flight_categories = ['Short-haul', 'Medium', 'Long-haul']
```

```python
# Create a list of categories
flight_categories = ["Short-haul", "Medium", "Long-haul"]

# Create short_flights
short_flights = "^0h|^1h|^2h|^3h|^4h"

# Create medium_flights
medium_flights = "^5h|^6h|^7h|^8h|^9h"

# Create long_flights
long_flights = "10h|11h|12h|13h|14h|15h|16h"
```

```python
import numpy as np

# Create conditions for values in flight_categories to be created
conditions = [
    (df2["Duration"].str.contains(short_flights)),
    (df2["Duration"].str.contains(medium_flights)),
    (df2["Duration"].str.contains(long_flights))
]

# Apply the conditions list to the flight_categories
df2["Duration_Category"] = np.select(conditions, 
                                        flight_categories,
                                        default="Extreme duration")

# Plot the counts of each category
sns.countplot(data=df2, x="Duration_Category")
plt.show()
```
